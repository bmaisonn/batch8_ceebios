# Data exploration

```python
important_columns = [
    "countryCode",
    "locality",
    'kingdom',
    'phylum', 'class', 'order', 'family', 'genus',
    "specificEpithet",
    "species",
    "genericName",
    "acceptedScientificName",
]
```

```python
for col in important_columns:
    data[col].value_counts().head(10).sort_values().plot(kind = "barh",figsize = (5,4),title = col)
    plt.show()
```
